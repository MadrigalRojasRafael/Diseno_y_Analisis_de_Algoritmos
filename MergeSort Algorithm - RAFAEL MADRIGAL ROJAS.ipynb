# **MergeSort Algorithm**

---

El **algoritmo MergeSort** es un **algoritmo de ordenamiento** que **utiliza** el **enfoque** de "**Divide y Vencerás**" para **ordenar elementos** en una **lista**.

El **MergeSort** es ampliamente **utilizado** en la práctica debido a su **eficiencia** y **estabilidad**. Es especialmente **útil** cuando se necesita **ordenar grandes conjuntos de datos**. Además, su **algoritmo** es **fácil de entender** y su **comportamiento predecible** lo convierten en una **elección** popular para **aplicaciones** que **requieren ordenamiento eficiente**.

### **Algoritmo MergeSort en Lenguaje Natural**

1. **Si** la **lista** tiene **tamaño 0** o **1**, entonces ya **está ordenada**. En ese caso, se **devuelve** la **lista sin cambios**.
2. De lo contrario, se **divide** la **lista** en **dos mitades aproximadamente iguales**.
3. Se **aplica recursivamente** el **algoritmo MergeSort** a **cada** una de las **mitades**. Esto implica **repetir** los **pasos 1 al 3 hasta** que cada **mitad** sea de **tamaño 0 o 1**.
4. Una vez que se han **ordenado** las **mitades**, se **combina** las **dos mitades** para **obtener** la **lista final ordenada**.
    1. Para **combinar** las **dos mitades**, se **toma** el **elemento más pequeño** de **cada mitad** y se **coloca** en una **nueva lista**.
    2. Se **repite** este **proceso hasta** que se hayan **agotado** los **elementos** en **ambas mitades**.
    3. **Si** aún **quedan elementos** en una de las **mitades**, se **agregan** a la **nueva lista en orden**.
5. **Finalmente**, se **devuelve** la **lista combinada** y **ordenada**.


```python
#Importamos el módulo numpy para crear la lista de prueba.
import numpy as np

#Creamos la función para implementar el algoritmo MergeSort.
def merge_sort(lista):
    if len(lista) <= 1:
        return lista

    # Dividir la lista en dos mitades
    mitad = len(lista) // 2
    mitad_izq = merge_sort(lista[:mitad])
    mitad_der = merge_sort(lista[mitad:])

    # Combinar las dos mitades ordenadas
    lista_ordenada = merge(mitad_izq, mitad_der)
    return lista_ordenada

def merge(izquierda, derecha):
    lista_combinada = []
    i = j = 0

    # Comparar y combinar los elementos de las dos mitades
    while i < len(izquierda) and j < len(derecha):
        if izquierda[i] <= derecha[j]:
            lista_combinada.append(izquierda[i])
            i += 1
        else:
            lista_combinada.append(derecha[j])
            j += 1

    # Agregar los elementos restantes de la mitad izquierda
    lista_combinada.extend(izquierda[i:])

    # Agregar los elementos restantes de la mitad derecha
    lista_combinada.extend(derecha[j:])

    return lista_combinada

#Ejemplo de uso del algoritmo Bubble Sort para ordenar una lista.
print(" Lista Desordenada ".center(125, "-"))
lista = np.random.randint(10, 100, size=100000)
print(lista)
print(" Lista Ordenada ".center(125, "-"))
print(f"{merge_sort(lista)}")
```

### **Complejidad Espacial y Temporal del Algoritmo MergeSort**

```python
import matplotlib.pyplot as plt
import time

def merge_sort(lista):
    if len(lista) <= 1:
        return lista

    mitad = len(lista) // 2
    mitad_izq = merge_sort(lista[:mitad])
    mitad_der = merge_sort(lista[mitad:])

    return merge(mitad_izq, mitad_der)

def merge(izquierda, derecha):
    lista_combinada = []
    i = j = 0

    while i < len(izquierda) and j < len(derecha):
        if izquierda[i] <= derecha[j]:
            lista_combinada.append(izquierda[i])
            i += 1
        else:
            lista_combinada.append(derecha[j])
            j += 1

    lista_combinada.extend(izquierda[i:])
    lista_combinada.extend(derecha[j:])

    return lista_combinada

def graficar_complejidad_temporal(n):
    tiempos = []

    for i in range(1, n + 1):
        lista = list(range(i, 0, -1))
        inicio = time.time()
        merge_sort(lista)
        fin = time.time()
        tiempo = fin - inicio
        tiempos.append(tiempo)

    plt.plot(range(1, n + 1), tiempos)
    plt.xlabel('Tamaño de la lista')
    plt.ylabel('Tiempo (segundos)')
    plt.title('Complejidad Temporal del MergeSort')
    plt.show()

def graficar_complejidad_espacial(n):
    espacios = []

    for i in range(1, n + 1):
        lista = list(range(i, 0, -1))
        memoria = lista.__sizeof__()
        espacios.append(memoria)

    plt.plot(range(1, n + 1), espacios)
    plt.xlabel('Tamaño de la lista')
    plt.ylabel('Espacio (bytes)')
    plt.title('Complejidad Espacial del MergeSort')
    plt.show()

# Ejemplo de uso
n = 10000  # Tamaño máximo de la lista
graficar_complejidad_temporal(n)
graficar_complejidad_espacial(n)
```
