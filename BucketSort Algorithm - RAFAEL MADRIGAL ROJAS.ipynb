# **BucketSort Algorithm**

---

El **Bucket Sort Algorithm** es un **algoritmo de ordenamiento** que **clasifica** los **elementos** de una **lista dividiéndolos** en un **número finito de** "**cubetas**". **Cada** "**cubeta**" se **ordena individualmente utilizando recursivamente** el **Bucket Sort**. Al **final**, los **elementos** de las "**cubetas**" se **concatenan** para **obtener** la **lista ordenada completa**.

El **Bucket Sort** es **útil** en **situaciones específicas** donde los **elementos** están **uniformemente distribuidos** en un **rango conocido**, se **recomienda** su **uso** en casos en los que las **condiciones** sean **favorables** para **obtener** un **rendimiento óptimo**, como cuando son **grandes conjuntos de datos** o un **número conocido de datos**.

---

### **Algortimo BucketSort en Lenguaje Natural**

1. **Selecciona** un **número fijo** de **cubetas vacías**.
2. **Recorre** la **lista de elementos** y **coloca cada elemento** en la **cubeta correspondiente** según un **criterio** de **asignación**. 
3. **Ordena individualmente cada cubeta** utilizando **Bucket Sort recursivamente**.
4. **Concatena** los **elementos ordenados** de **cada cubeta** en **orden** para **obtener** la **lista ordenada final**.


```python
#Importamos el módulo numpy para crear la lista de prueba.
import numpy as np

#Creamos la función para implementar el algoritmo Bucket Sort.
import numpy as np

def bucket_sort(lista):
    if len(lista) == 0:
        return lista

    # Encontrar el valor máximo y mínimo en la lista
    min_valor = np.min(lista)
    max_valor = np.max(lista)

    # Calcular el rango de cada cubeta
    rango = (max_valor - min_valor) / len(lista)

    # Crear las cubetas utilizando listas vacías
    cubetas = [[] for _ in range(len(lista))]

    # Asignar elementos a las cubetas
    for elemento in lista:
        index = int((elemento - min_valor) / rango)
        # Manejar el caso cuando el índice está fuera del rango válido de cubetas
        if index >= len(cubetas):
            index = len(cubetas) - 1
        cubetas[index].append(elemento)

    # Ordenar individualmente cada cubeta utilizando NumPy
    for i in range(len(cubetas)):
        cubetas[i] = np.sort(cubetas[i])

    # Concatenar las cubetas ordenadas
    lista_ordenada = np.concatenate(cubetas)

    return lista_ordenada.astype(int).tolist()

#Ejemplo de uso del algoritmo Bucket Sort para ordenar una lista.
print(" Lista Desordenada ".center(125, "-"))
lista = np.random.randint(1, 100, size=100)
print(lista)
print(" Lista Ordenada ".center(125, "-"))
print(f"{bucket_sort(lista)}")
```

### **Complejidad Espacial y Temporal del Algoritmo Bubble Sort** 

```python
import matplotlib.pyplot as plt
import random
import time

def bucket_sort(lista):
    # Crear el número de cubetas adecuado
    num_cubetas = len(lista)
    cubetas = [[] for _ in range(num_cubetas)]

    # Asignar elementos a las cubetas
    for elemento in lista:
        indice = int(elemento * num_cubetas)
        cubetas[indice].append(elemento)

    # Ordenar cada cubeta individualmente
    for cubeta in cubetas:
        cubeta.sort()

    # Concatenar las cubetas ordenadas
    lista_ordenada = []
    for cubeta in cubetas:
        lista_ordenada.extend(cubeta)

    return lista_ordenada

def graficar_complejidad_temporal(n):
    tiempos = []

    for i in range(1, n + 1):
        lista = [random.random() for _ in range(i)]
        inicio = time.time()
        bucket_sort(lista)
        fin = time.time()
        tiempo = fin - inicio
        tiempos.append(tiempo)

    plt.plot(range(1, n + 1), tiempos)
    plt.xlabel('Tamaño de la lista')
    plt.ylabel('Tiempo (segundos)')
    plt.title('Complejidad Temporal del Bucket Sort')
    plt.show()

def graficar_complejidad_espacial(n):
    espacios = []

    for i in range(1, n + 1):
        lista = [random.random() for _ in range(i)]
        espacio = lista.__sizeof__()
        espacios.append(espacio)

    plt.plot(range(1, n + 1), espacios)
    plt.xlabel('Tamaño de la lista')
    plt.ylabel('Espacio (bytes)')
    plt.title('Complejidad Espacial del Bucket Sort')
    plt.show()

# Ejemplo de uso
n = 10000  # Tamaño máximo de la lista
graficar_complejidad_temporal(n)
graficar_complejidad_espacial(n)
```
