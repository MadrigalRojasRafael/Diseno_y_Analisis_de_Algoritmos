# **Bubble Sort Algorithm**

---

El **algoritmo de ordenamiento Bubble Sort** es un **algoritmo simple**, comúnmente **utilizado** para **ordenar elementos** en una **lista**. 

El **algoritmo** toma su **nombre** de la forma en que los **elementos** "**burbujean**" a medida que **se comparan** y **se intercambian repetidamente durante** el **proceso** de **ordenamiento**.   

### **Algortimo Bubble Sort en Lenguaje Natural**
1. **Comenzando** desde el **principio** de la **lista**, **compara** el **primer elemento** **con** el **siguiente elemento**.
2. **Si** el **primer elemento** es **mayor que** el **siguiente**, **intercámbialos**.
3. **Avanza** al **siguiente** **par** de **elementos adyacentes** y **repite** el **paso 2**.
4. **Continúa recorriendo** la **lista** y **realizando comparaciones** e **intercambios adyacentes** **hasta** el **final** de la **lista**.
5. Al **llegar** al **final** de la **lista**, el **elemento más grande** estará en su **posición correcta** en la **parte derecha** de la **lista**.
6. **Repite** los **pasos** del **1 al 5** para los **elementos restantes** de la **lista**, **excepto** el **último elemento**.
7. **Repite** los **pasos** del **1 al 6** **hasta** que **no** se **realicen intercambios** en un **recorrido completo** de la **lista**.

```python
#Importamos el módulo numpy para crear la lista de prueba.
import numpy as np

#Creamos la función para implementar el algoritmo Bubble Sort.
def bubble_sort(lista):
    n = len(lista)

    # Realizar recorridos hasta que no se realicen intercambios
    for i in range(n - 1):
        intercambios = False

        # Realizar comparaciones e intercambios adyacentes
        for j in range(n - 1 - i):
            if lista[j] > lista[j + 1]:
                # Intercambiar elementos
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
                intercambios = True

        # Si no se realizaron intercambios en el recorrido, la lista está ordenada
        if not intercambios:
            break

    return lista

#Ejemplo de uso del algoritmo Bubble Sort para ordenar una lista.
print(" Lista Desordenada ".center(125, "-"))
lista = np.random.randint(1, 100, size=1000)
print(lista)
print(" Lista Ordenada ".center(125, "-"))
print(f"{bubble_sort(lista)}")
```

### **Complejidad Espacial y Temporal del Algoritmo Bubble Sort** 

```python
import matplotlib.pyplot as plt
import time

def bubble_sort(lista):
    n = len(lista)

    for i in range(n - 1):
        intercambios = False

        for j in range(n - 1 - i):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
                intercambios = True

        if not intercambios:
            break

def graficar_complejidad_temporal(n):
    tiempos = []

    for i in range(1, n + 1):
        lista = list(range(i, 0, -1))  # Generar una lista inversa de tamaño i
        inicio = time.time()
        bubble_sort(lista)
        fin = time.time()
        tiempo = fin - inicio
        tiempos.append(tiempo)

    plt.plot(range(1, n + 1), tiempos)
    plt.xlabel('Tamaño de la lista')
    plt.ylabel('Tiempo (segundos)')
    plt.title('Complejidad Temporal del Bubble Sort')
    plt.show()

def graficar_complejidad_espacial(n):
    espacios = []

    for i in range(1, n + 1):
        lista = list(range(i, 0, -1))
        memoria = lista.__sizeof__()  # Obtener el tamaño en memoria de la lista
        espacios.append(memoria)

    plt.plot(range(1, n + 1), espacios)
    plt.xlabel('Tamaño de la lista')
    plt.ylabel('Espacio (bytes)')
    plt.title('Complejidad Espacial del Bubble Sort')
    plt.show()

# Ejemplo de uso
n = 1000  # Tamaño máximo de la lista
graficar_complejidad_temporal(n)
graficar_complejidad_espacial(n)
```
