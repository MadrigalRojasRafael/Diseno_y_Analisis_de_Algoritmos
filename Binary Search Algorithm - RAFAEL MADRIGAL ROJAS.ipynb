# **Binary Search Algorithm**

---

El algoritmo de búsqueda binaria es un algoritmo eficiente para buscar un elemento en una lista ordenada. 

### **Algoritmo Binary Search en Lenguaje Natural**

1. Establecer los límites izquierdo y derecho de la lista: 
   1. izquierda = 0
   2. derecha = longitud de la lista - 1
2. Mientras la izquierda sea menor o igual a la derecha:
   1. Calcular el índice medio:
      - medio = (izquierda + derecha) // 2
   2. Si el valor en el índice medio es igual al valor buscado:
      - Devolver el índice medio
   3. Si el valor en el índice medio es menor que el valor buscado:
      - Establecer el nuevo límite izquierdo en medio + 1
   d. Si el valor en el índice medio es mayor que el valor buscado:
      - Establecer el nuevo límite derecho en medio - 1
3. Si no se encuentra el valor buscado, devolver -1.

```python
#Importamos el módulo numpy para crear la lista de prueba.
import numpy as np

#Creamos la función para implementar el algoritmo Binary Search.
def binary_search(lista, valor):
    izquierda = 0
    derecha = len(lista) - 1

    while izquierda <= derecha:
        medio = (izquierda + derecha) // 2

        if lista[medio] == valor:
            return medio
        elif lista[medio] < valor:
            izquierda = medio + 1
        else:
            derecha = medio - 1

    return -1

#Ejemplo de uso del algoritmo Binary Search para buscar un elemento dentro de una lista.
print(" Parámetros ".center(125, "-"))
print("Lista:")
lista = np.random.choice(range(1, 1001), size=1000, replace=False)
print(lista)
valor_buscado = 352
print(f"Valor Buscado: {valor_buscado}")
print(" Resultado ".center(125, "-"))
resultado = binary_search(lista, valor_buscado)
if resultado != -1:
    print(f"El valor {valor_buscado} se encuentra en la posición {resultado}.")
else:
    print(f"El valor {valor_buscado} no se encuentra en la lista.")
```
