# **QuickSort Algorithm**

---

El **algoritmo QuickSort** es un **algoritmo de ordenamiento** que **utiliza** la estrategia de "**divide y vencerás**". Se basa en la **elección** de un **elemento pivote** en la **lista** y luego **divide** la **lista** en **dos subconjuntos**: **uno** con **elementos menores que** el **pivote** y **otro** con **elementos mayores**. Luego, **aplica recursivamente** el **mismo proceso** en los **subconjuntos** **hasta** que la **lista** esté completamente **ordenada**.

---

### **Algoritmo QuickSort en Lenguaje Natural**

1. **Si** la **lista** **tiene** un **tamaño menor o igual a 1**, entonces ya **está ordenada**. En ese caso, se **devuelve** la **lista sin cambios**.
2. **De lo contrario**, se **elige** un **elemento** de la lista como **pivote**. Este **pivote** puede ser **cualquier elemento** de la **lista**, pero **generalmente** se **elige** el **primer elemento**.
3. **Crear dos subconjuntos vacíos**: **uno** para los **elementos menores o iguales al pivote** y **otro** para los **elementos mayores al pivote**.
4. **Recorrer** la **lista**, **excluyendo** el **pivote**, y se **coloca cada elemento** **en** el **subconjunto correspondiente**, ya sea el de los menores/iguales o el de los mayores.
5. Se **aplica recursivamente** el **algoritmo QuickSort** a **cada** uno de los **subconjuntos** (**repetir** los **pasos** del **1 al 4** **hasta** que los **subconjuntos** sean de **tamaño 0 o 1**).
6. **Finalmente**, se **combina** el **subconjunto** de los **menores/iguales**, el **pivote** **y** el **subconjunto de los mayores** para **obtener** la **lista ordenada completa**.

```python
#Importamos el módulo numpy para crear la lista de prueba.
import numpy as np

#Creamos la función para implementar el algoritmo QuickSort.
def quicksort(lista):
    if len(lista) <= 1:
        return lista
    else:
        pivote = lista[0]
        menores = [elemento for elemento in lista[1:] if elemento <= pivote]
        mayores = [elemento for elemento in lista[1:] if elemento > pivote]
        return quicksort(menores) + [pivote] + quicksort(mayores)

#Ejemplo de uso del algoritmo QuickSort para ordenar una lista.
print(" Lista Desordenada ".center(125, "-"))
lista = np.random.randint(1, 100, size=100000)
print(lista)
print(" Lista Ordenada ".center(125, "-"))
print(f"{quicksort(lista)}")
```

### **Complejidad Espacial y Temporal del Algoritmo Quicksort**

```python
import matplotlib.pyplot as plt
import time

def quicksort(lista):
    if len(lista) <= 1:
        return lista
    else:
        pivote = lista[0]
        menores = [elemento for elemento in lista[1:] if elemento <= pivote]
        mayores = [elemento for elemento in lista[1:] if elemento > pivote]
        return quicksort(menores) + [pivote] + quicksort(mayores)

def graficar_complejidad_temporal(n):
    tiempos = []

    for i in range(1, n + 1):
        lista = list(range(i, 0, -1))  # Generar una lista inversa de tamaño i
        inicio = time.time()
        quicksort(lista)
        fin = time.time()
        tiempo = fin - inicio
        tiempos.append(tiempo)

    plt.plot(range(1, n + 1), tiempos)
    plt.xlabel('Tamaño de la lista')
    plt.ylabel('Tiempo (segundos)')
    plt.title('Complejidad Temporal del QuickSort')
    plt.show()

def graficar_complejidad_espacial(n):
    espacios = []

    for i in range(1, n + 1):
        lista = list(range(i, 0, -1))
        memoria = lista.__sizeof__()  # Obtener el tamaño en memoria de la lista
        espacios.append(memoria)

    plt.plot(range(1, n + 1), espacios)
    plt.xlabel('Tamaño de la lista')
    plt.ylabel('Espacio (bytes)')
    plt.title('Complejidad Espacial del QuickSort')
    plt.show()

# Ejemplo de uso
n = 2000  # Tamaño máximo de la lista
graficar_complejidad_temporal(n)
graficar_complejidad_espacial(n)
```
