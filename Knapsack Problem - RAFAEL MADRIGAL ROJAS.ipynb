# **Knapsack Problem**

---

### **Descripción del Problema:**
Se tiene una **mochila** con **capacidad máxima W**.
Y se tiene un **conjunto de objetos O** que está **compuesto por objetos**:
    
~~~
oi = { pi , bi } 
Dónde:
    pi → peso del objeto i.
    bi → beneficio de llevar el objeto i en la mochila.
~~~

**Para cualquier conjunto de objetos O**, **encontrar** un **algoritmo** que siempre de la **configuración con beneficio máximo** que puedo llevar con la **capacidad** de la **mochila**.

---

### **Algoritmo Voraz**

+ Los **algoritmos voraces (greedy algorithms)**, son algoritmos que **resuelven problemas** mediante la **toma de decisiones locales óptimas** en cada paso, **sin considerar** las posibles **consecuencias futuras** de esas decisiones. 

+ En otras palabras, **en cada paso** del algoritmo, **se toma la mejor decisión posible** basada en la información disponible en ese momento, con la esperanza de que **esa decisión conduzca** a una **solución óptima global**.

#### **Algoritmo Voraz en Lenguaje Natural**

1. **Calcular** la **relación valor-peso** para cada elemento de la lista de objetos.
2. **Ordenar** los **objetos** en **orden descendente** según su **relación valor-peso**.
3. **Inicializar** una **lista** vacía para **almacenar** la **solución**.
4. **Inicializar** el **peso total** y **valor total** de la solución en **0**.
5. **Iterar** sobre los **objetos** en **orden descendente** según su **relación valor-peso**:
    1. **Si** el **peso** del **objeto más el peso total de la solución** es **menor o igual** a la **capacidad** de la **mochila**, **añadir** el **objeto** a la **solución** y **actualizar** el **peso total** y **valor total** de la **solución**.
    2. **Si** el **peso total** de la **solución excede** la **capacidad** de la **mochila**, **detener** la **iteración**.
6. **Devolver** la **lista** de **objetos** seleccionados en la **solución** y el **valor total obtenido**.


```python
#Importamos el módulo numpy para manejar de manera fácil los elementos de las listas.
import numpy as np

#Creamos la función para implementar el algoritmo voraz para resolver el Knapsack Problem.
def knapsack_voraz(pesos, valores, capacidad):
    objetos = [(pesos[i], valores[i]) for i in range(len(pesos))]
    objetos.sort()
    mochila = []
    peso_total = 0
    valor_total = 0
    
    for objeto in objetos:
        if peso_total + objeto[0] <= capacidad:
            mochila.append(objeto)
            peso_total += objeto[0]
            valor_total += objeto[1]
        resultado = f"Mochila: {mochila}\nPeso Total: {peso_total}\nValor Total: {valor_total}"
    return resultado

#Ejemplo de uso del algoritmo Voraz para resolver el Knapsack Problem.
print(" Parámetros ".center(125, "-"))
pesos=np.random.randint(1, 50, size=10000)
print(f"Pesos: {pesos}")
valores=np.random.randint(1, 10, size=10000)
print(f"Valores: {valores}")
capacidad = 850
print(f"Capacidad: {capacidad}")

print(" Resultado ".center(125, "-"))
ejemploVoraz = knapsack_voraz(pesos, valores, capacidad)
print(ejemploVoraz)
```

### **Complejidad Temporal y Espacial del Algoritmo Voraz para el Knapsack Problem**

```python
import time
import sys
import numpy as np
import matplotlib.pyplot as plt

def knapsack_voraz(pesos, valores, capacidad):
    objetos = [(pesos[i], valores[i]) for i in range(len(pesos))]
    objetos.sort()
    mochila = []
    peso_total = 0
    valor_total = 0
    
    for objeto in objetos:
        if peso_total + objeto[0] <= capacidad:
            mochila.append(objeto)
            peso_total += objeto[0]
            valor_total += objeto[1]
        resultado = f"Mochila: {mochila}\nPeso Total: {peso_total}\nValor Total: {valor_total}"
    return resultado

# Generar datos de prueba
tamanos = np.arange(1000, 20001, 1000)
tiempos = []
espacios = []

for n in tamanos:
    pesos = np.random.randint(1, 50, size=n)
    valores = np.random.randint(1, 10, size=n)
    capacidad = 850
    
    # Medir tiempo de ejecución
    tiempo_inicio = time.time()
    knapsack_voraz(pesos, valores, capacidad)
    tiempo_fin = time.time()
    tiempo_ejecucion = tiempo_fin - tiempo_inicio
    tiempos.append(tiempo_ejecucion)

    # Medir espacio utilizado
    espacio_utilizado = sys.getsizeof(pesos) + sys.getsizeof(valores)
    espacios.append(espacio_utilizado)

# Graficar tiempos de ejecución
plt.plot(tamanos, tiempos, marker='o')
plt.xlabel('Tamaño del problema (n)')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.title('Complejidad temporal del algoritmo voraz para el Knapsack Problem')
plt.show()

# Graficar espacio utilizado
plt.plot(tamanos, espacios, marker='o')
plt.xlabel('Tamaño del problema (n)')
plt.ylabel('Espacio utilizado (bytes)')
plt.title('Complejidad espacial del algoritmo voraz para el Knapsack Problem')
plt.show()
```

### **Programación Dinámica**

+ La **programación dinámica** es una **técnica** utilizada en informática y matemáticas para **resolver problemas complejos dividiéndolos** en otros **más pequeños** y **resolviendo** cada problema, **guardando** sus **resultados** y **reutilizándolos** en lugar de leerlos nuevamente.  

+ La **programación dinámica** ofrece **beneficios** como la **optimización de tiempo**, la **garantía de solución óptima**, la **reutilización de resultados**, la **versatilidad** y la **escalabilidad**. Estos beneficios hacen que sea una **técnica** valiosa para **resolver problemas computacionales** de manera **eficiente** y **efectiva**.

#### **Algoritmo de Programación Dinámica en Lenguaje Natural**

1. **Crear** una **matriz bidimensional** con **N+1 filas** y **W+1 columnas**, donde **N** es el **número de elementos disponibles** y **W** es la **capacidad máxima de la mochila**. 
2. **Inicializar** la **primera fila** y la **primera columna** de la tabla **con ceros**, ya que representan el caso base donde no hay elementos seleccionados o no hay capacidad en la mochila.
3. **Recorrer** los **elementos disponibles** uno por uno **en un bucle**. **Para cada elemento**, consideramos dos opciones: **incluirlo** o **excluirlo** en la **solución**.
4. **Para cada capacidad posible de la mochila**, **desde 0 hasta W**, **calcular** el **valor máximo** que se puede obtener **considerando** el **elemento actual**.
5. **Si** el **peso del elemento actual** es **menor o igual** a la **capacidad de la mochila** actual, comparamos dos casos:
    1. **Incluir** el **elemento actual**: **sumando su valor al valor máximo obtenido** para la **capacidad restante** de la **mochila**.
    2. **Excluir** el **elemento actual**: **tomando el valor máximo obtenido** hasta ahora para la misma **capacidad de la mochila sin considerar** el **elemento actual**.
    3. **Seleccionar** el **valor máximo** entre estas dos opciones y **almacenarlo** en la **matriz**.
6. **Si** el **peso del elemento actual** es **mayor que** la **capacidad de la mochila actual**, **no incluirlo** y simplemente **tomar** el **valor máximo obtenido** hasta ahora **para** la **misma capacidad de la mochila sin considerar** el **elemento actual**.
7. **Después** de **completar** el **bucle** y **llenar** toda la **tabla**, el **valor máximo alcanzado** se encuentra en la **celda tabla[N][W]**, donde **N** es el **número total de elementos** y **W** es la **capacidad máxima de la mochila**.
8. Para **reconstruir** la **solución óptima**, **comenzando** desde la **celda tabla[N][W]** y **retrocediendo hacia** la **celda tabla[1][1]**.
9. En **cada paso**, **comparar** el **valor en la celda actual con el valor en la celda superior**. Si son **diferentes**, significa que el **elemento correspondiente se incluyó en la solución óptima**. **Agregar** ese **elemento** a nuestra **lista de solución** y **pasar** a la **celda superior considerando** el **peso** y el **valor restantes**.
10. **Repetir** este **proceso hasta** llegar a **la celda tabla[1][1]**. La **lista de elementos agregada** a la **solución** es la **combinación óptima** que **maximiza el valor total** dentro de la **capacidad de la mochila**.
11. **Devolver** la **solución** obtenida.


```python
#Importamos el módulo numpy para manejar de manera fácil los elementos de las listas.
import numpy

#Creamos la función para implementar el algoritmo de Programación Dinámica para resolver el Knapsack Problem.
def knapsack_dp(capacidad, pesos, valores):
    n = len(pesos)

    matriz = np.zeros((n + 1, capacidad + 1))

    for i in range(1, n + 1):
        for w in range(1, capacidad + 1):
            if pesos[i - 1] <= w:
                matriz[i][w] = max(valores[i - 1] + matriz[i - 1][w - pesos[i - 1]], matriz[i - 1][w])
            else:
                matriz[i][w] = matriz[i - 1][w]

    solucion = []
    i = n
    w = capacidad
    while i > 0 and w > 0:
        if matriz[i][w] != matriz[i - 1][w]:
            solucion.append((pesos[i - 1], valores[i - 1]))
            w -= pesos[i - 1]
        i -= 1

    peso_total = sum(peso for peso, _ in solucion)
    valor_total = sum(valor for _, valor in solucion)
    
    resultado = f"Mochila: {solucion}\nPeso Total: {peso_total}\nValor Total: {valor_total}"
    return resultado

#Ejemplo de uso del algoritmo de Programación Dinámica para resolver el Knapsack Problem.
print(" Parámetros ".center(125, "-"))
print(f"Pesos: {pesos}")
print(f"Valores: {valores}")
print(f"Capacidad: {capacidad}")
print(" Resultado ".center(125, "-"))
ejemploDynammicProgramming = knapsack_dp(capacidad, pesos, valores)
print(ejemploDynammicProgramming)
```

### **Complejidad Temporal y Espacial del Algoritmo de Programación Dinámica para el Knapsack Problem**

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import time

def knapsack(capacidad, pesos, valores):
    n = len(pesos)

    matriz = np.zeros((n + 1, capacidad + 1))

    for i in range(1, n + 1):
        for w in range(1, capacidad + 1):
            if pesos[i - 1] <= w:
                matriz[i][w] = max(valores[i - 1] + matriz[i - 1][w - pesos[i - 1]], matriz[i - 1][w])
            else:
                matriz[i][w] = matriz[i - 1][w]

    solucion = []
    i = n
    w = capacidad
    while i > 0 and w > 0:
        if matriz[i][w] != matriz[i - 1][w]:
            solucion.append((pesos[i - 1], valores[i - 1]))
            w -= pesos[i - 1]
        i -= 1

    peso_total = sum(peso for peso, _ in solucion)
    valor_total = sum(valor for _, valor in solucion)
    
    resultado = f"Mochila: {solucion}\nPeso Total: {peso_total}\nValor Total: {valor_total}"
    return resultado

# Generar datos para la gráfica
tamanos = np.arange(1000, 20001, 1000)
tiempos = []
espacios = []

for n in tamanos:
    pesos = np.random.randint(1, 100, size=n)
    valores = np.random.randint(1, 100, size=n)
    capacidad = 850

    # Medir tiempo de ejecución
    tiempo_inicio = time.time()
    knapsack(capacidad, pesos, valores)
    tiempo_fin = time.time()
    tiempo_ejecucion = tiempo_fin - tiempo_inicio
    tiempos.append(tiempo_ejecucion)

    # Medir espacio utilizado
    espacio_utilizado = sys.getsizeof(pesos) + sys.getsizeof(valores)
    espacios.append(espacio_utilizado)

# Graficar complejidad temporal
plt.plot(tamanos, tiempos, marker='o')
plt.xlabel('Tamaño del problema (n)')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.title('Complejidad temporal del algoritmo de programación dinámica para el Knapsack Problem')
plt.show()

# Graficar complejidad espacial
plt.plot(tamanos, espacios, marker='o')
plt.xlabel('Tamaño del problema (n)')
plt.ylabel('Espacio utilizado (bytes)')
plt.title('Complejidad espacial del algoritmo de programación dinámica para el Knapsack Problem')
plt.show()
```
